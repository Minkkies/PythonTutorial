### 3. จงเขียนโปรแกรมเครื่องมือตรวจคําตอบข้อสอบปรนัย โดยโปรแกรมมีคุณสมบัติดังต่อไปนี้
- รับสตริงเฉลยและสตริงคําตอบนักเรียนต่อเนื่อง
- เมื่อผู้ใช้พิมพ์ 'exit' (ไม่สนตัวพิมพ์เล็ก-ใหญ่) ในช่องสตริงเฉลย
โปรแกรมจะจบการทํางานแล้วแสดงข้อความว่า 'Exiting program.
Goodbye!'
- เปรียบเทียบคําตอบแต่ละข้อแบบไม่สนตัวพิมพ์เล็ก - ใหญ่ (เช่น 'A' เท่ากับ 'a')
แล้วนับจํานวนข้อที่ถูก แสดงข้อความว่า "Number of correct
answers: [จํานวนข้อที่ถูก]"
- ถ้าจํานวนข้อของเฉลยไม่ตรงกับของคําตอบ ให้แสดง Incomplete answer

**ตัวอย่าง**
```
--- Multiple Choice Answer Checker ---
Enter the string of correct answers: ABABC
Enter the string of student's answers: ABBBC
Number of correct answers: 4
Enter the correct answers (type 'exit' to quit): XYZ
Enter the student's answers: xy
Incomplete answer
Enter the correct answers (type 'exit' to quit): exit
Exiting program. Goodbye!
```

#### นับจำนวนข้อถูก

`count = sum(1 for a, s in zip(ans.upper(), std.upper()) if a == s)`

เขียนเต็มได้แบบนี้ใช้ Method ที่มีมาทำ
```
count = 0
  for a, s in zip(ans.upper(), std.upper()):
    if a == s:
      count += 1
```

````python
print("--- Multiple Choice Answer Checker ---")

first = True  # ไว้เช็คว่ารอบแรกไหม

while True:

  if first:
    ans = input("Enter the string of correct answers: ")
    first = False
  else:
    ans = input("Enter the correct answers (type 'exit' to quit): ")

  # เช็ค exit (ไม่สนตัวพิมพ์เล็ก-ใหญ่)
  if ans.lower() == "exit":
    print("Exiting program. Goodbye!")
    break

  # รับคำตอบนักเรียน
  if first:
    std = input("Enter the string of student's answers: ")
  else:
    std = input("Enter the student's answers: ")

  # ความยาวไม่เท่ากัน
  if len(ans) != len(std):
    print("Incomplete answer")
    continue

  # นับจำนวนข้อถูก พร้อมปรับให้ทั้ง 2 ค่าที่รับมาเป็นตัวใหญ่ทั้งหมด
  count = 0
  for i in range(len(ans)):
    if ans[i].upper() == std[i].upper():
      count += 1

  print(f"Number of correct answers: {count}")
````
