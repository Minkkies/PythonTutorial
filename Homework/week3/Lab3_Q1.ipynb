# Q1:
  
###จงเขียนโปรแกรมเพื่อตรวจสอบว่าวงกลมสองวงมีการ ทับซ้อน (overlap) แตะกัน(touch) หรืออยู่แยกกัน (free) โดยรับข้อมูลของวงกลม 2 วง บรรทัดละหนึ่งวงประกอบด้วยจํานวนจริง 3 จํานวนคั่นด้วยช่องว่าง แทนพิกัด x กับ y ของจุดศูนย์กลางและรัศมีของวงกลม แล้วตรวจว่าวงกลมสองวงที่รับมาทับกันหรือแตะกันหรือไม่ให้แสดงคําว่า touch เมื่อขอบของทั้งสองวงแตะกันพอดี แสดงคําว่า overlap เมื่อสองวงทับกัน ถ้าไม่แตะหรือทับ ให้แสดงคําว่า free

**Hint**
1. หาระยะห่าง คํานวณระยะห่างระหว่างจุดศูนย์กลางของวงกลมทั้งสองจุด (x1, y1) และ
(x2, y2) โดยใช้สูตร Euclidean Distance d = !(x! − x")" + (y! − y")"
2. หาผลรวมรัศมีนํารัศมีของทั้งสองวงมาบวกกัน (r1+ r2)
3. เปรียบเทียบ นําค่าระยะห่าง (d) มาเทียบกับผลรวมรัศมี (r1 + r2)
• ถ้า d = r1 + r2 ขอบชนกันพอดี (touch)
• ถ้า d < r1 + r2 วงกลมซ้อนทับกัน (overlap)
• ถ้า d > r1 + r2 อยู่ห่างกัน (free)

**ตัวอย่าง**
```
---Circle Overlap Detector---
Enter x, y, and radius for Circle 1: 0 0 4
Enter x, y, and radius for Circle 2: 0 6 3
overlap

---Circle Overlap Detector---
Enter x, y, and radius for Circle 1: 0 0 2
Enter x, y, and radius for Circle 2: 0 4 2
touch
---Circle Overlap Detector---
Enter x, y, and radius for Circle 1: 0 0 9
Enter x, y, and radius for Circle 2: 0 12 2
Free
```

````python
import math

x1, y1, radius1 = map(float, input('Enter x, y, and radius for Circle 1: ').split())
# map คือ แปลงเป็น int โดยตรง
x2, y2, radius2 = map(float, input('Enter x, y, and radius for Circle 2: ').split())

d = math.sqrt(pow((x1 - x2), 2) + pow(y1 - y2, 2))

if d == radius1 + radius2:
    print('Touch')
elif d > radius1 + radius2:
    print('Free')
else:
  print('Overlap')
````
