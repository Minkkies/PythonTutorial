แล้วคํานวณนํ้ามันที่ใช้ตามอัตราการใช้นํ้ามันเฉลี่ยในตารางScreenshot 2025-12-09 154726.png

**ตัวอย่าง**
```
Enter travel distance (km.): 100
Enter the type of car: 1
Enter driving pattern: 2
Fuel required: 10.00 litres.

Enter travel distance (km.): 150
Enter the type of car: 3
Enter driving pattern: 1
Fuel required: 16.67 litres
```

````python
import math

distance = float(input('Enter travel distance (km.): '))
car = float(input('Enter the type of car: '))
pattern = float(input('Enter driving pattern: '))
fuel = 0.0


if car == 1:
    if pattern == 1:
        fuel = distance / 12
    elif pattern == 2:
        fuel = distance / 10

elif car == 2:
    if pattern == 1:
        fuel = distance / 10
    elif pattern == 2:
        fuel = distance / 8

elif car == 3:
    if pattern == 1:
        fuel = distance / 9
    elif pattern == 2:
        fuel = distance / 7

print(f'Fuel required: {fuel} litres')
````
