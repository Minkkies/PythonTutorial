## Q2 จงเขียนโปรแกรมที่วนรับชื่อผู้ใช้จนกว่าจะเป็น Lisa หรือใส่ชื่อผิดครบ 3 ครั้ง

### ตัวอย่างการทำงาน:

Enter username : **Lisa** <br>
Hello, Lisa

### ตัวอย่างการทำงาน:

Enter username : **test** <br>
Incorrect. Enter again : **12345**  <br>
Incorrect. Enter again : **Lisa** <br>
Hello, Lisa

### ตัวอย่างการทำงาน:

Enter username : **test**  <br>
Incorrect. Enter again : **12345**  <br>
Incorrect. Enter again : **my name** <br>
Not allowed. Incorrect name.

```python
username = input("Enter username : ")
count = 0
while username !='Lisa' and count < 2:
  username = input("Incorrect. Enter again : ")
  count += 1

if username !='Lisa' :
  print('Not allowed. Incorrect name.')
else:
  print(f"Hello, {username}")
```
