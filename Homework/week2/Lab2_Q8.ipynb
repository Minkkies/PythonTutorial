# Q8 จงเขียนโปรแกรมที่วนรับจำนวนเงินที่ต้องการเบิกจากบัญชี โดยจะยอมให้เบิกถ้ายังมีเงินในบัญชีพอ และวนรับจำนวนเงินที่ต้องการเบิกไปจนเงินในบัญชีหมดพอดี (กำหนดให้เริ่มต้นมีเงินในบัญชี 50,000 บาท)


### ตัวอย่างการทำงาน:

withdraw : **50000** <br>
Balance is 0

### ตัวอย่างการทำงาน:

withdraw : **45000**  <br>
withdraw : **5000**  <br>
Balance is 0

### ตัวอย่างการทำงาน:

withdraw : **60000**  <br>
Insufficient fund. <br>
withdraw : **40000**  <br>
withdraw : **20000**  <br>
Insufficient fund.<br>
withdraw : **10000**  <br>
Balance is 0

````python
balance = 50000

print(f"You have money in bank {balance} baht")

while balance > 0:
    withdraw = int(input("withdraw :"))

    if withdraw > balance:
        print("Insufficient fund.")
    else:
        balance -= withdraw

print(f"Balance is {balance} baht")
````

แบบย่อ (one-liner):
- ถ้าเงื่อนไขเป็นจริง → ทำการคำนวณ (balance -= withdraw)
- ถ้าเงื่อนไขเป็นเท็จ → แค่ print("Insufficient fund.")

ปัญหาคือ Python ไม่อนุญาตให้ใช้ print() ใน ternary expression กับการคำนวณ เพราะ print() คืนค่า None

**วิธีแก้**

เราต้องแยกเป็น if–else ปกติ หรือใช้ lambda / function เล็ก ๆ

1️⃣ วิธีปกติ (อ่านง่ายที่สุด)
```python
if withdraw <= balance:
    balance -= withdraw
else:
    print("Insufficient fund.")
```
2️⃣ ถ้าอยากทำเป็นแบบย่อด้วย lambda
```python
def withdraw_money(balance, withdraw):
    return balance - withdraw if withdraw <= balance else print("Insufficient fund.") or balance

balance = 50000
withdraw = int(input("withdraw: "))
balance = withdraw_money(balance, withdraw)
```
- ใช้ or balance เพราะ print() คืนค่า None
- ทำให้ผลลัพธ์ไม่ error
