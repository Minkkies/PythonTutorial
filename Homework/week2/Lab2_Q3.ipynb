## Q3 จงเขียนโปรแกรมที่รับจำนวนนักเรียนที่เข้าสอบ แล้ววนรับคะแนนของนักเรียนที่เข้าสอบให้ครบและจงหาผลรวมคะแนน

### ตัวอย่างการทำงาน:

Number of students : **5** <br>
Student 1 : **10** <br>
Student 2 : **6.5** <br>
Student 3 : **0** <br>
Student 4 : **9** <br>
Student 5 : **8.25** <br>
Total score : 33.75 <br>

### ตัวอย่างการทำงาน:

Number of students : **-1** <br>
Incorrect Input. Number of students : **7** <br>
Student 1 : **10** <br>
Student 2 : **6.5** <br>
Student 3 : **0** <br>
Student 4 : **9** <br>
Student 5 : **8.25** <br>
Student 6 : **9.50** <br>
Student 7 : **7.25** <br>
Total score : 50.5 <br>

```python
student = int(input("Number of students : "))
while student <= 0:
    student = int(input("Incorrect Input. Number of students : "))

total = 0
for i in range(student):
    score = float(input(f"Student {i+1}: "))
    total += score

print(f"Total score: {total}")
```

โค้ดนี้จะ error เพราะ for i in student: ทำไม่ได้ — student เป็นตัวเลข (int) ไม่ใช่ลิสต์หรือ range

**อธิบายสั้น ๆ**
- range(student) จะสร้างลำดับตัวเลขตั้งแต่ 0 ถึง student−1
- ทำให้ใช้ i+1 เพื่อแสดงลำดับนักเรียนได้ถูกต้อง
