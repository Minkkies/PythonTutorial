## Q4 จงเขียนโปรแกรมที่รับจำนวนนักเรียนที่เข้าสอบ (เป็นจำนวนบวกเท่านั้น) แล้ววนรับคะแนนของนักเรียนที่เข้าสอบให้ครบและหา
- คะแนนเฉลี่ยของนักเรียนทั้งหมด
- คะแนนเฉลี่ยของนักเรียนที่สอบผ่าน (ได้คะแนนอย่างน้อย 5)
- คะแนนเฉลี่ยของนักเรียนที่สอบไม่ผ่าน (ได้คะแนนต่ำกว่า 5)
- คะแนนสูงสุดของนักเรียนทั้งหมด

### ตัวอย่างการทำงาน:

Number of students : **3** <br>
Student 1 : **10** <br>
Student 2 : **6.5** <br>
Student 3 : **0** <br>
Average score : 5.5 <br>
Average passing score : 8.25 <br>
Average failing score : 0.0 <br>
Highest score : 10.0 <br>

### ตัวอย่างการทำงาน:

Number of students : **-1** <br>
Incorrect Input. Number of students : **7** <br>
Student 1 : **10** <br>
Student 2 : **6.5** <br>
Student 3 : **0** <br>
Student 4 : **9** <br>
Student 5 : **8.25** <br>
Student 6 : **9.50** <br>
Student 7 : **7.25** <br>
Average score : 7.214285714285714 <br>
Average passing score : 8.416666666666666 <br>
Average failing score : 0.0 <br>
Highest score : 10.0 <br>

### ตัวอย่างการทำงาน:

Number of students : **3** <br>
Student 1 : **1** <br>
Student 2 : **2.5** <br>
Student 3 : **3.5** <br>
Average score : 2.3333333333333335 <br>
Average passing score : No student passed <br>
Average failing score : 2.3333333333333335 <br>
Highest score : 3.5 <br>

### ตัวอย่างการทำงาน:

Number of students : **3** <br>
Student 1 : **6** <br>
Student 2 : **5.5** <br>
Student 3 : **9.8** <br>
Average score : 7.1000000000000005 <br>
Average passing score : 7.1000000000000005 <br>
Average failing score : No student failed <br>
Highest score : 9.8 <br>

```python
student = int(input("Number of students : "))
while student <= 0:
    student = int(input("Incorrect Input. Number of students : "))

total = 0.0
max_score = -1 # เก็บค่าสูงสุดเริ่มต้น
max_student = 0 # เก็บลำดับของนักเรียนที่คะแนนสูงสุด
count_pass = 0
count_fail = 0
score_pass = 0.0
score_fail = 0.0

for i in range(student):
    score = float(input(f"Student {i+1}: "))
    total += score

    # หา คะแนนเฉลี่ยของนักเรียนที่สอบผ่านและไม่ผ่าน
    if score >= 5:
      count_pass +=1
      score_pass += score
    else:
      count_fail +=1
      score_fail += score

    # หา highest score
    if score > max_score:
        max_score = score
        max_student = i + 1

print(f"Average score: {total / student}")

# ค่าเฉลี่ยนักเรียนที่ผ่าน => example of conditional expression
print(f'Average passing score : {score_pass / count_pass}') if count_pass > 0 else print('Average passing score : No student passed')

# ค่าเฉลี่ยนักเรียนที่ไม่ผ่าน
if count_fail > 0:
    print(f'Average failing score : {score_fail / count_fail}')
else:
    print('Average failing score : No student failed')

print(f"Highest score: {max_score} (Student {max_student})")
```
