```python
import math

a, b, c = input('Enter coefficients a, b, c : ').split(",")
a = float(a)
b = float(b)
c = float(c)

x = (-b + math.sqrt(b*b - 4*a*c)) / (2*a)
x1 = (-b - math.sqrt(b*b - 4*a*c)) / (2*a)
print(f"x = {x},{x1}")
```
